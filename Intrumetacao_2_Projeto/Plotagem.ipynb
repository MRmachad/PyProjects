```python
import csv
import pandas as pd
from pandas import *
import numpy as np
import matplotlib.pyplot as plt
from pylab import stem, title, xlabel, ylabel, subplot, grid, ylim, subplots


dados = []
dados2 = []
dados_tempADC = []
dados_tensaoADC = []
dados_tempESP = []
dados_tensaoESP = []
numero_de_leituras = 211 #Número de leituras que você quer plotar

# Pegando os dados
df = pd.read_csv('Tensao_ad.csv')
dados = df['dados'].values.tolist()
df2 = pd.read_csv('Tensao_esp.csv')
dados2 = df2['dados'].values.tolist()

# Arrumando os valores de tensão e temperatura
aux = 0
while (aux < numero_de_leituras):
    dados_tempADC.append(float(dados[aux])*40)
    dados_tempESP.append(float(dados2[aux])*40)
    aux = aux + 1

aux = 0
while (aux < numero_de_leituras):
    dados_tensaoADC.append(float(dados[aux]))
    dados_tensaoESP.append(float(dados2[aux]))
    aux = aux + 1

# Linha de "tempo"
x = np.arange(0, 1 * numero_de_leituras, 1)

# Plotando o grágifo
# Separadamente
fig, (ax, ax1) = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Gráficos de temperatura')
ax.plot(x, dados_tempADC)
ax.set_ylabel('Temperatura ADS')
ax1.plot(x, dados_tempESP)
ax1.set_ylabel('Temperatura ESP')

# Juntos
fig1, ax2 = plt.subplots(figsize=(12, 5))
fig1.suptitle('Comparação das Temperaturas')
ax3 = ax2.twinx()
ax2.plot(x, dados_tempADC, 'g-')
ax2.set_ylabel('Temperatura ADS', color='g')
ax3.plot(x, dados_tempESP, 'b-')
ax3.set_ylabel('Temperatura ESP', color='b')

plt.show()
```
